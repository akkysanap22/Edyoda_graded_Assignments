# 2. Find all houses which has more than 2 bedroom and price less than 50k

```python
house_data[(house_data.Bedroom > 2) & (house_data.Price < 50000)].head()
```

# 3. Find all houses which has more than 3 bedroom and price less than 50k. Update their price with 60k

```python
house_data[(house_data.Bedroom > 3) & (house_data.Price < 50000)].head()
```

# Updating price with 60K

```python
def f(r):
    if r.Bedroom > 3 and r.Price < 50000:
        return 60000
    else:
        return r.Price
house_data['New_Price'] = house_data.apply(f,axis = 1)
```

```python
house_data[house_data.Bedroom > 3].head()
```

# 8. If gender is male and age is less than 35 then change diabetes result to NO .

```python

def diabetes(d):
    if d.gender == 'Male' & d.Age < 35:
        return 'NO'
    else:
        return d.diabetes
    
    
health_data['diabetes'] = health_data.apply(diabetes,axis = 1)
```
